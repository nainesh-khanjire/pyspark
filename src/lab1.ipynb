```python
conf = SparkConf().setAppName('lab1').setMaster('local[*]')
sc = SparkContext(conf=conf)
```

```python
sc
```

```python
# this is a list of numbers
l = [1,2,3,4,5,6]
```

```python
# created RDD with 4 partitions , data will be divided into 4 parts
lrdd = sc.parallelize(l,4)
```

```python
print(type(lrdd))
```

```python
print(lrdd)
```

```python
# print RDD using collect 
# this is distributed across workers so collect and print it
lrdd.collect()
```

```python
#create RDD ffrom a tuple
l2 = (1,2,3,4,5,6)
tuplerdd = sc.parallelize(l2,4)
tuplerdd.collect()
```

```python
# create RDD from dictionary
d = {'a':1,'b':2,'c':3,'d':4}
dictrdd = sc.parallelize(d,4)
dictrdd.collect()
```

## *create RDD from existing RDD*

```python
r = dictrdd.map(lambda x:x*2)
```
